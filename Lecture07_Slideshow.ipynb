#PHYSICS 91SI: Plotting!

Plotting is one of the most important aspects of data analysis. Unfortunately, your hard work means nothing if your data isn't presented aesthetically.

Below, I cover some basics of plotting:
- subplots, sizes, colors, opacity, styles
- legends, labels, colormaps, colorbars
- text, order-of-plotting

Of course, this is by no means a comprehensive guide. With that in mind, let's learn about plotting!

```python
%matplotlib inline

import __future__ #because i still have python 2.5
import numpy as np
import matplotlib.pyplot as plt
```

Here's the standard line graph that you guys already know and love.

```python
x = np.arange(20)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

You can plot more than one equation at the same time. Don't forget to label the axes!

```python
y1 = np.cos(x)
plt.plot(x,y,x,y1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

You can specify the thickness of the lines via the Line2D properties.<br>
(More at http://matplotlib.org/api/lines_api.html#matplotlib.lines.Line2D)

```python
plt.plot(x,y,lw=3)
plt.plot(x,y1,lw=7)
plt.show()
```

You can also specify the order in which the lines are plotted.<br>
(More at http://matplotlib.org/examples/pylab_examples/zorder_demo.html)

```python
plt.plot(x,y,lw=3,zorder=2)
plt.plot(x,y1,lw=7,zorder=1)
plt.show()
```

There's more! You can also specify the colors and styles of the lines via the Line2D properties.

```python
plt.plot(x,y,color='red',linestyle='dotted')
plt.plot(x,y1,color='blue',linestyle='dashed')
plt.show()
```

Warning: after you put in a "keyword arg," you can't put in a non-keyword argument.

```python
plt.plot(x,y,color='red',linestyle='dotted',x,y1,color='blue',linestyle='dashed')
```

There are certain shorcuts in matplotlib.pyplot for plotting color and style.

```python
plt.plot(x,y,'r:',x,y1,'b--')
plt.show()
```

By specifying a "marker" style, you can emphasize your data points.

```python
plt.plot(x,y,marker='o')
plt.show()
```

But by specifying only a marker style using the shortcut... you get something that looks like a scatter plot!

```python
plt.plot(x,y,'o')
plt.show()
```

Here's another way to get a scatter plot-looking plot:

```python
plt.plot(x,y,marker='v',linestyle='none')
plt.show()
```

And here's the actual way to get a scatter plot:

```python
plt.scatter(x,y,marker='v')
plt.show()
```

You can also give your data points error bars.<br>
(More at http://matplotlib.org/1.2.1/examples/pylab_examples/errorbar_demo.html and http://matplotlib.org/examples/statistics/errorbar_limits.html)

```python
plt.scatter(x,y,marker='v')
plt.errorbar(x,y,xerr=0.1,yerr=0.05,ecolor='red',fmt='none')
plt.show()
```

You can combine shortcuts for colors and markers for plt.plot:

```python
plt.plot(x,y,'ro',x,y1,'bv')
plt.show()
```

You can't use shortcuts for plt.scatter which means each scatter plot deserves its own line of code.

```python
plt.scatter(x,y,color='r',marker='o')
plt.scatter(x,y1,color='b',marker='v')
plt.show()
```

Did the above look confusing? well, you can put multiple plots on the same set of axes.

```python
plt.scatter(x,y,color='r',marker='o')
plt.scatter(x,y1,color='b',marker='v')
plt.plot(x,y,'r:',x,y1,'b--')
plt.show()
```

You can also put multiple plots on different subplots.<br>
Let's also set titles to follow the graphs easier :)

```python
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(x,y,color='r')
axarr[0].set_xlim([0,10])
axarr[1].plot(x,y1,color='b')
axarr[0].set_title('sharing the x axis')
plt.show()
```

Subplots come in all different shapes and sizes.

```python
f, (ax0,ax1,ax2) = plt.subplots(3, sharey=True)
ax0.plot(x,y,color='r')
ax0.set_ylim([0,2])
ax1.plot(x,y1,color='b')
ax2.plot(y,y1,color='g')
f.subplots_adjust(hspace=0)
ax0.set_title('sharing the y axis')
plt.show()
```

Want to get fancier?<br>
More at http://matplotlib.org/examples/pylab_examples/subplots_demo.html

```python
f, ((ax0, ax1),(ax2, ax3)) = plt.subplots(2,2)
ax0.set_title('solid line')
ax0.plot(x,y,'r-')
ax1.set_title('dashed line')
ax1.plot(x,y,'b--')
ax2.set_title('dashed-dotted line')
ax2.plot(x,y,'g-.')
ax3.set_title('dotted line')
ax3.plot(x,y,'m:')
plt.tight_layout()
plt.suptitle('the four different line styles', fontsize=20) #supertitle
plt.subplots_adjust(top=0.85)
plt.show()
```

Back to putting everything on the same set of axes, for simplicity's sake!<br>
You can specify the plot limits:

```python
plt.scatter(x,y,color='r',marker='o')
plt.scatter(x,y1,color='b',marker='v')
plt.plot(x,y,'r:',x,y1,'b--')
plt.xlim([-10,30])
plt.ylim([-2,2])
plt.show()
```

Forgot which was sin(x) and which was cos(x)?<br>
Let's label the plots and include a legend!

```python
plt.scatter(x,y,color='r',marker='o',label='sin(x)')
plt.scatter(x,y1,color='b',marker='v',label='cos(x)')
plt.plot(x,y,'r:',x,y1,'b--')
plt.xlim([-10,30])
plt.ylim([-2,2])
plt.legend()
plt.show()
```

Be careful when you label, though:

```python
plt.scatter(x,y,color='r',marker='o')
plt.scatter(x,y1,color='b',marker='v')
plt.plot(x,y,'r:',x,y1,'b--',label='sin(x) and/or cos(x)')
plt.xlim([-10,30])
plt.ylim([-2,2])
plt.legend()
plt.show()
```

Want to learn more about legends?<br>
More at http://matplotlib.org/users/legend_guide.html

```python
plt.scatter(x,y,color='r',marker='o',label='sin(x)')
plt.scatter(x,y1,color='b',marker='v',label='cos(x)')
plt.plot(x,y,'r:',x,y1,'b--')
plt.plot([-10,-5,0,5,10,20],[0.3,0.3,0.3,0.3,0.3,0.3],'ko',label='a random line')
plt.xlim([-10,20])
plt.ylim([-2,1])
plt.legend(numpoints=4,scatterpoints=5,loc='best',fontsize='20',frameon=True,shadow=True)
plt.show()
```

Back to scatter plots!

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
plt.show()
```

You can give these random data points random pretty colors.<br>
Ideally, you would have three arrays of the same length:<br>
<tab>xdata, ydata, colordata</tab><br><br>
This way, you can plot 3 parameters on a 2D graph!

```python
colors = np.random.rand(50)
plt.scatter(x,y,c=colors)
plt.show()
```

Want to plot 4 parameters on a 2D graph?<br>
Change the sizes of the data points!<br>
<tab>xdata, ydata, colordata, sizedata

```python
area = np.pi * (15 * np.random.rand(50))**2  # 0 to 15 point radiuses
plt.scatter(x,y,c=colors,s=area)
plt.show()
```

We can make the above graph look prettier...

... By fiddling with the transparencies and outlines!

```python
plt.scatter(x,y,c=colors,s=area,edgecolors='none',alpha=0.5)
plt.show()
```

"alpha" is the opacity keyword argument.<br>
Opacity is uniform across all data points, and thus the alpha parameter does not accept an array.

```python
plt.scatter(x,y,c=colors,s=area,edgecolors='none',alpha=0.2)
plt.show()
```

Don't like random colors? Use a colormap!<br>
(More at http://matplotlib.org/examples/color/colormaps_reference.html)

```python
plt.scatter(x,y,c=colors,cmap=plt.cm.gist_rainbow,s=area,edgecolors='none',alpha=0.5)
plt.show()
```

You can include a colorbar, too:

```python
h = plt.scatter(x,y,c=colors,cmap=plt.cm.gist_rainbow,s=area,edgecolors='none',alpha=0.5)
h = plt.colorbar()
plt.show()
```

It is up to you to determine what color and size mean on your graph.<br>
Below, size correlates to x values, while color correlates to y values.

```python
size2 = np.pi * (15 * x)**2
colors2 = y
h = plt.scatter(x,y,c=colors2,cmap=plt.cm.gist_rainbow,s=size2,edgecolors='none',alpha=0.5)
h = plt.colorbar()
plt.show()
```

You can put text on your plot as well, though hopefully more tastefully than I have.

```python
plt.scatter(x,y,c=colors2,cmap=plt.cm.gist_rainbow,s=size2,edgecolors='none',alpha=0.5)
plt.text(0,0,'origin')
plt.text(0.4,0.4,'hi!',ha='right')
plt.text(0.4,0.4,'hi?',ha='left',color='magenta')
plt.text(max(x),max(y),'max x,y val',fontsize=13,va='bottom',rotation=180)
plt.text(max(x),max(y),'max x,y val',fontsize=13,va='top',color='blue')
plt.show()
```

And very, very briefly: histograms!<br>
(More at http://matplotlib.org/1.2.1/examples/pylab_examples/histogram_demo.html)

```python
xhist1 = np.random.randn(10000)
xhist2 = xhist1 * 2
plt.hist(xhist1,50,color='green',normed=1)
plt.hist(xhist2,50,color='blue',edgecolor='none',normed=1,alpha=0.5)
plt.xlim([-8,8])
plt.show()
```
